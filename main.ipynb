# "What Makes a YouTube Video Go Viral?"

##  Objective:
Analyze the Trending YouTube Videos Dataset to uncover patterns and features (tags, views, likes, categories, etc.) that make a video trend or go viral.

```python
df = pd.read_csv("CAvideos.csv")
df.head()
#df.tail()
```

```python
print("shape:" , df.shape)
print("columns:" , df.columns.tolist())
df.head()
```

# Cleaning of Data

```python
#convert publish time to datetime
df['publish_time'] = pd.to_datetime(df['publish_time'])

#fill missing tags
df['tags'] = df['tags'].fillna("")

#remove rows with disabled links
df = df[(df['comments_disabled'] == False) & (df['ratings_disabled'] == False)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(x='views' , y= 'likes',data = df)
plt.title("views vs likes")
plt.xlabel("views")
plt.ylabel("likes")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='views' , y= 'likes',data = df)
plt.title("views vs likes")
plt.xlabel("views")
plt.ylabel("likes")
plt.grid(True)
plt.show()
```

# Category Distribution

```python
print(df.columns)
```

```python
df['category_id'].value_counts().plot(kind='bar', color ='orange',figsize=(8,4))
plt.title("Trending video categories in canada")
plt.xlabel("category id")
plt.ylabel("count")
plt.show()
```

# Like-to-View ratio (Virality Indicator)

```python
df['like_view_ratio'] = df['likes'] / df['likes']
top_virals = df.sort_values(by='like_view_ratio', ascending = False).head(10)
print(top_virals[['title' , 'channel_title','like_view_ratio']])
```

# WordCloud for Tags

```python
from wordcloud import WordCloud

tags_combined =" ".join(df['tags'].tolist()).replace("|" ," " )

wordcloud = WordCloud(width=800, height=400 , background_color ='white' ).generate(tags_combined)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title("top tags in canadian trending videos")
plt.show()
```

# Top Trending channel

```python
df['channel_title'].value_counts().head(10).plot(kind ='bar' , color = 'green')
plt.title("Top 10 channels by number of trending videos")
plt.ylabel("video count")
plt.xticks(rotation=45)
plt.show()
```

# Trending video publish times

```python
df['hour'] = df['publish_time'].dt.hour
df['hour'].value_counts().sort_index().plot(kind = 'bar',color = 'purple')
plt.title("trending video publish hours")
plt.xlabel("hour of day")
plt.ylabel("number of videos")
plt.show()
```
